```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f= open('WhatsApp Chat with CS 8A (Official).txt', 'r', encoding='utf-8')
```

```python
data= f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
message = re.split(pattern, data)
message
```

```python
message = re.split(pattern, data)[1:]
message
```

```python
len(message)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
dataf = pd.DataFrame({'user_messages':message, 'messages_dates':dates})
```

```python
# convert type of message dates into datetime bcZ they are in string format
dataf['messages_dates'] = pd.to_datetime(dataf['messages_dates'], format= '%d/%m/%Y, %H:%M - ')

dataf.rename(columns={'messages_dates': 'date'}, inplace=True)
```

# Basic Analysis

```python
dataf.isnull().sum()
```

```python
# Messages ko clean karna (extra spaces hatana)
dataf['user_messages'] = dataf['user_messages'].str.strip()
```

```python
total_messages = dataf.shape[0]
print(f"Total Messages: {total_messages}")
```

```python
# User names extract karna (messages mein jo ":" se pehle hai)
dataf['user'] = dataf['user_messages'].str.extract(r'^(.*?):')[0]
unique_users = dataf['user'].dropna().unique()
print(f"Unique Users: {len(unique_users)}")
```

```python
dataf.head(20)
```

```python
dataf.dropna(inplace=True)
dataf.head(10)
```

```python
dataf['Year'] = dataf['date'].dt.year
dataf['Month'] = dataf['date'].dt.month_name()  # Month extract karna
dataf['Day'] = dataf['date'].dt.date # Sirf date extract karna
dataf['hour'] = dataf['date'].dt.hour  # Hour extract karna
dataf['minutes'] = dataf['date'].dt.minute
```

```python
dataf.head()
```

```python
daily_messages = dataf.groupby('Day').size()

# Daily messages ka line chart
plt.figure(figsize=(10, 5))
daily_messages.plot(kind='line', color='blue')
plt.title('Daily Messages')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.show()
```

```python
# Hourly messages

hourly_messages = dataf.groupby('hour').size()

# Bar chart
plt.figure(figsize=(10, 5))
hourly_messages.plot(kind='bar', color='orange')
plt.title('Messages Sent Per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Messages')
plt.xticks(range(24), [f'{i}:00' for i in range(24)], rotation=45)
plt.show()
```

```python
# Group data by month
monthly_messages = dataf.groupby('Month').size()

# Bar chart
plt.figure(figsize=(10, 5))
monthly_messages.plot(kind='bar', color='brown')
plt.title('Messages Sent Per Month')
plt.xlabel('Month of the Year')
plt.ylabel('Number of Messages')

# Set the x-ticks to display month names
plt.xticks(ticks=range(len(monthly_messages)), labels=monthly_messages.index, rotation=45)
plt.show()
```

```python
# Group data by year
yearly_messages = dataf.groupby('Year').size()

# Bar chart
plt.figure(figsize=(10, 5))
yearly_messages.plot(kind='bar', color='orange')
plt.title('Messages Sent Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Messages')

# Set the x-ticks to display years
plt.xticks(ticks=range(len(yearly_messages)), labels=yearly_messages.index, rotation=45)
plt.show()
```

```python
busiest_day = daily_messages.idxmax()
print(f"Busiest Day: {busiest_day}")
```

```python
active_user = dataf['user'].value_counts().idxmax()
print(f"Most Active User: {active_user}")
```

```python
# Top 10 most active users
top_users = dataf['user'].value_counts().head(10)

# Bar chart
plt.figure(figsize=(10, 5))
top_users.plot(kind='bar', color='green')
plt.title('Top 10 Most Active Users')
plt.xlabel('User')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import Counter
from wordcloud import WordCloud

# Step 3: Words Extract Karna
# Sab messages ko ek string mein combine karna
all_messages = ' '.join(dataf['user_messages'].dropna()).lower()

# Words extract karna (alphabets aur numbers ke words)
words = re.findall(r'\b\w+\b', all_messages)

# Step 4: Words Ki Frequency Count Karna
word_counts = Counter(words)

# # WordCloud generate karna
# wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_messages)

# Step 5: Word Cloud Banana
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)
# Word Cloud ko display karna
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Most Used Words')
plt.show()
```

```python
from textblob import TextBlob

# Sentiment analyze karna
def get_sentiment(text):
    analysis = TextBlob(str(text))
    return analysis.sentiment.polarity  # -1 (negative) se 1 (positive) tak

dataf['sentiment'] = dataf['user_messages'].apply(get_sentiment)

# Positive, negative, neutral messages count karna
dataf['sentiment_type'] = dataf['sentiment'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))
sentiment_counts = dataf['sentiment_type'].value_counts()
print(sentiment_counts)
```

```python
# Sentiment counts
sentiment_counts = dataf['sentiment_type'].value_counts()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=['green', 'red', 'blue'])
plt.title('Sentiment Analysis')
plt.show()
```

```python
# Specific keywords count karna
keywords = ['javed', 'quiz', 'assignment']
keyword_counts = {keyword: dataf['user_messages'].str.contains(keyword, case=False).sum() for keyword in keywords}
print(keyword_counts)
```
